```python
import pandas as pd
import numpy as np
```

```python
hop_file = "../pandashomeworks/purchase_data.csv"

hop_data_df = pd.read_csv(hop_file)
hop_data_df.head()
```

```python
total_players = hop_data_df.groupby(['SN'])
player_count = total_players['SN'].nunique().sum()
player_count_df = pd.DataFrame({"Players": player_count}, index=[0])
player_count_df.head()
```

```python
item_count = hop_data_df.groupby(['Item ID'])

unique_item_count = item_count['Item ID'].nunique().sum()

avg_price = hop_data_df['Price'].mean()

num_purch = hop_data_df['Item ID'].count()

tot_rev = hop_data_df['Price'].sum()

purch_df = pd.DataFrame({"Number of Unique Items": unique_item_count, "Average Price": avg_price, 
                         "Number of Purchases": num_purch, "Total Revenue": tot_rev}, index=[0])
purch_df['Total Revenue'] = purch_df['Total Revenue'].map('${:,.2f}'.format)
purch_df['Average Price'] = purch_df['Average Price'].map('${:,.2f}'.format)
purch_df.head()
```

```python
gender_df = hop_data_df[["SN", "Gender"]]
gender_dup_drop = gender_df.drop_duplicates()
gender_dup_drop.head()
```

```python
male_count = gender_dup_drop.loc[gender_dup_drop["Gender"] == "Male"].nunique().sum()
female_count = gender_dup_drop.loc[gender_dup_drop["Gender"] == "Female"].nunique().sum()
other_count = gender_dup_drop.loc[gender_dup_drop["Gender"] == "Other / Non-Disclosed"].nunique().sum()
male_percent = male_count / 576 * 100
female_percent = female_count / 576 * 100
other_percent = other_count / 576 * 100

gender_demo_df = pd.DataFrame({"Gender": ("Male", "Female", "Other / Non-Disclosed"),
                               "Total Count": (male_count, female_count, other_count), 
                               "Percentage of Players": (male_percent, female_percent, other_percent)})
gender_demo_df['Percentage of Players'] = gender_demo_df['Percentage of Players'].map('{:,.2f}%'.format)
gender_demo_df.head()
```

```python
#Gender count
gender_count_tot = hop_data_df.groupby('Gender').count()['SN']

#Average purchse Price
gender_avg_price = hop_data_df.groupby('Gender').mean()['Price']


#Total Purchase Value
gender_purch_tot = hop_data_df.groupby('Gender').sum()['Price']

#Average total purchase per person
gender_avgpurch_per = gender_purch_tot / gender_count_tot

gender_purch_df = pd.DataFrame({"Purchase Count": gender_count_tot, 
                                "Average Purchase Price": gender_avg_price,
                                "Total Purchase": gender_purch_tot,
                                "Average Total Purchase per Person": gender_avgpurch_per })

gender_purch_df['Average Purchase Price'] = gender_purch_df['Average Purchase Price'].map('${:,.2f}'.format)
gender_purch_df['Total Purchase'] = gender_purch_df['Total Purchase'].map('${:,.2f}'.format)
gender_purch_df['Average Total Purchase per Person'] = gender_purch_df['Average Total Purchase per Person'].map('${:,.2f}'.format)

gender_purch_df.head()
```

```python
age_bins = [5, 10, 15, 20, 25, 30, 35, 40, 45]
bin_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
pd.cut(hop_data_df["Age"], age_bins, labels=bin_labels).head()
```

```python
hop_data_df["Age Group"] = pd.cut(hop_data_df["Age"], age_bins, labels=bin_labels)
hop_data_df.head()
```

```python
ages_grp = hop_data_df.groupby("Age Group").count()["Age"]
avg_ages = ages_grp / 576 * 100

age_df = pd.DataFrame({"Total Count": ages_grp, 
                       "Percantage of Players": avg_ages})
age_df['Percantage of Players'] = age_df['Percantage of Players'].map('{:,.2f}%'.format)
age_df.head(8)
```

```python
age_purch_count =  hop_data_df.groupby('Age Group').count()["Item ID"]
age_avg_price = hop_data_df.groupby('Age Group').mean()['Price']
age_totpurch_price = hop_data_df.groupby('Age Group').sum()['Price']
age_avgpurh_per = age_totpurch_price / age_purch_count

age_purch_df = pd.DataFrame({"Purchase Count": age_purch_count, 
                             "Average Purchase Price": age_avg_price,
                             "Total Purchase Value": age_totpurch_price,
                             "Average Total Purchase per Person": age_avgpurh_per})

age_purch_df['Average Purchase Price'] = age_purch_df['Average Purchase Price'].map('${:,.2f}'.format)
age_purch_df['Total Purchase Value'] = age_purch_df['Total Purchase Value'].map('${:,.2f}'.format)
age_purch_df['Average Total Purchase per Person'] = age_purch_df['Average Total Purchase per Person'].map('${:,.2f}'.format)

age_purch_df.head(8)
```

```python
top_purch_count =  hop_data_df.groupby('SN').count()["Item ID"]
age_avg_price = hop_data_df.groupby('Age Group').mean()['Price']
age_totpurch_price = hop_data_df.groupby('Age Group').sum()['Price']
age_avgpurh_per = age_totpurch_price / age_purch_count

age_purch_df = pd.DataFrame({"Purchase Count": age_purch_count, 
                             "Average Purchase Price": age_avg_price,
                             "Total Purchase Value": age_totpurch_price,
                             "Average Total Purchase per Person": age_avgpurh_per})

age_purch_df['Average Purchase Price'] = age_purch_df['Average Purchase Price'].map('${:,.2f}'.format)
age_purch_df['Total Purchase Value'] = age_purch_df['Total Purchase Value'].map('${:,.2f}'.format)
age_purch_df['Average Total Purchase per Person'] = age_purch_df['Average Total Purchase per Person'].map('${:,.2f}'.format)

age_purch_df.head(8)
```

```python
spender_count = hop_data_df.groupby("SN").count()["Purchase ID"] 
spender_avg = hop_data_df.groupby("SN").mean()['Price']
spender_totpurch = hop_data_df.groupby("SN").sum()['Price']

spender_df = pd.DataFrame({"Purchase Count": spender_count, 
                             "Average Purchase Price": spender_avg,
                             "Total Purchase Value": spender_totpurch})

spender_df['Average Purchase Price'] = spender_df['Average Purchase Price'].map('${:,.2f}'.format)
spender_df['Total Purchase Value'] = spender_df['Total Purchase Value'].map('${:,.2f}'.format)
fspender_df = spender_df.sort_values("Purchase Count", ascending=False)
fspender_df.head()
```

```python
item_names = hop_data_df.groupby("Item ID").count()["Purchase ID"]
item_price = hop_data_df.groupby('Item Name').mean()["Price"]
itotpurch_val = item_price * item_price

popitems_df = pd.DataFrame({"Purchase Count": item_names, 
                            "Item Price": item_price,
                            "Total Purchase Value": itotpurch_val})

popitems_df['Item Price'] = popitems_df['Item Price'].map('${:,.2f}'.format)
popitems_df['Total Purchase Value'] = popitems_df['Total Purchase Value'].map('${:,.2f}'.format)
fpopitems_df = popitems_df.sort_values("Purchase Count", ascending=False)
fpopitems_df.head()
```
